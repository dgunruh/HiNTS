```python
root = Tk()
foldername = fd.askdirectory()
root.withdraw()

subdirectories = [x[0] for x in os.walk(foldername)]
allDiameterData = []
for n,directory in enumerate(subdirectories):
    dList = []
    for file in os.listdir(directory):
        if file[-4:]==".inp":
            f = open(directory + "\\" + file,'r')
            index = 1


            for line in f:
                if index > 6:
                    newline = line.split(', ')
                    dList.append(float(newline[3]))
                index += 1
            f.close()

    dArray = np.asarray(dList)
    #print(dArray)
    mean = np.mean(dArray)
    sigma = np.std(dArray)
    print("Mean is: ",mean, " and  is: ",sigma)
    allDiameterData.append([mean,sigma])
print(allDiameterData)
```
