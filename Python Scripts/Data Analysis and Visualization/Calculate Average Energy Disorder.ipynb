```python
import matplotlib.pyplot as plt
import numpy as np
from tkinter import Tk
from tkinter import filedialog as fd
import os
from scipy.optimize import curve_fit
from scipy.optimize import brute
from scipy import asarray as ar,exp
import math
```

```python
ha = 4.35974394e-18
ev = 1.602176487e-19
autoev = ha/ev
rytoev = autoev/2.0
```

```python
def conductionEnergy(diameter):
    return -4.238383 + 4.28063*(diameter**(-1.87117))
```

```python
def gaus(x,a,x0,sigma):
    return a*exp(-(x-x0)**2/(2*sigma**2))
```

```python
root = Tk()
folder_selected = fd.askdirectory(title = "Choose the folder containing the results")
#f = open(filename,'r')
root.withdraw()

ebulk = 22.0
rAverage = 6.5*10**(-9)/2
ll = 0.5*10**-9
enp = 1 + (ebulk - 1)/(1 + 18.05/((rAverage*10**9)**1.8))
em = 2.0
kappa = 2.0
f = 4/3*math.pi*(rAverage)**3/(8*(rAverage + ll)**3)
esolid = em*(ebulk*(1 + kappa*f) - em*(kappa*f - kappa))/(em*(kappa + f) + ebulk*(1 - f))
e0 = 8.85*10**(-12)
k = 1/(4*math.pi*e0)
e = 1.6*10**(-19)
cbedisorder = []
sigma0disorder = []
sigmadisorder = []
for file in os.listdir(folder_selected):
    cbeList = []
    sigma0List = []
    sigmaList = []
    i = 1
    if file[-4:] == '.inp':
        f = open(folder_selected + '/' + file)
        for line in f:
            if i > 6:
                newline = line.split()
                diameter = float(newline[3])
                cbe = conductionEnergy(diameter)
                r = diameter*10**(-9)/2
                sigma0 = (e**2*k/(2*r)*(1/esolid - 1/ebulk) + 0.47*e**2*k/(ebulk*r)*((ebulk - esolid)/(ebulk + esolid)))/(1.6*10**-19)
                sigma = (e**2*k/(r)*(1/esolid + 0.79/ebulk))/(1.6*10**-19)
                print(sigma)
                cbeList.append(cbe)
                sigma0List.append(sigma0)
                sigmaList.append(sigma)
            i += 1
            
        #now create histogram
        cbeArray = np.asarray(cbeList)
        hist, bins = np.histogram(cbeArray, bins = 'fd')
        centers = [(i + bins[n])/2 for n, i in enumerate(bins[1:])]
        n_bins = len(centers)
        mean = np.mean(cbeArray)
        sigma = np.std(cbeArray)
        popt,pcov = curve_fit(gaus,centers,hist,p0=[1,mean,sigma])
#         plt.plot(centers,hist,'b+:',label='data')
#         plt.plot(centers,gaus(centers,*popt),'ro:',label='gaussian fit')
#         plt.legend()
#         plt.grid()
#         plt.xlabel('cb energy (eV)')
#         plt.ylabel('Count')
#         plt.title('Gaussian Fit of Transformed Data')
#         plt.show()
        gaus_sigma = popt[2]
        fwhm = 2*math.sqrt(2*math.log(2))*gaus_sigma
        cbedisorder.append(abs(fwhm))
        
        #now create histogram
        sigma0Array = np.asarray(sigma0List)
        hist, bins = np.histogram(sigma0Array, bins = 'fd')
        centers = [(i + bins[n])/2 for n, i in enumerate(bins[1:])]
        n_bins = len(centers)
        mean = np.mean(sigma0Array)
        sigma = np.std(sigma0Array)
        popt,pcov = curve_fit(gaus,centers,hist,p0=[1,mean,sigma])
#         plt.plot(centers,hist,'b+:',label='data')
#         plt.plot(centers,gaus(centers,*popt),'ro:',label='gaussian fit')
#         plt.legend()
#         plt.grid()
#         plt.xlabel('cb energy (eV)')
#         plt.ylabel('Count')
#         plt.title('Gaussian Fit of Transformed Data')
#         plt.show()
        gaus_sigma = popt[2]
        fwhm = 2*math.sqrt(2*math.log(2))*gaus_sigma
        sigma0disorder.append(abs(fwhm))
        
        #now create histogram
        sigmaArray = np.asarray(sigmaList)
        hist, bins = np.histogram(sigmaArray, bins = 'fd')
        centers = [(i + bins[n])/2 for n, i in enumerate(bins[1:])]
        n_bins = len(centers)
        mean = np.mean(sigmaArray)
        sigma = np.std(sigmaArray)
        popt,pcov = curve_fit(gaus,centers,hist,p0=[1,mean,sigma])
#         plt.plot(centers,hist,'b+:',label='data')
#         plt.plot(centers,gaus(centers,*popt),'ro:',label='gaussian fit')
#         plt.legend()
#         plt.grid()
#         plt.xlabel('cb energy (eV)')
#         plt.ylabel('Count')
#         plt.title('Gaussian Fit of Transformed Data')
#         plt.show()
        gaus_sigma = popt[2]
        fwhm = 2*math.sqrt(2*math.log(2))*gaus_sigma
        sigmadisorder.append(abs(fwhm))
        
cbedisorderArray = np.asarray(cbedisorder)
meancbeDisorder = np.mean(cbedisorderArray)
print("mean cbe disorder is: ",meancbeDisorder," eV")

sigma0disorderArray = np.asarray(sigma0disorder)
meansigma0Disorder = np.mean(sigma0disorderArray)
print("mean sigma0 disorder is: ",meansigma0Disorder," eV")

sigmadisorderArray = np.asarray(sigmadisorder)
meansigmaDisorder = np.mean(sigmadisorderArray)
print("mean sigma disorder is: ",meansigmaDisorder," eV")

totalDisorder = meancbeDisorder + meansigma0Disorder + meansigmaDisorder
print("total disorder is: ",totalDisorder*1000," meV")
```
